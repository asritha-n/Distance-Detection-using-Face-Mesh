```python
import cv2
import cvzone 
from cvzone.FaceMeshModule import FaceMeshDetector
```

```python
#d = W*f/w where d is the distance, W is the width in cm and the w is the width in pixels, then f is the focal length
#we have average distance between the eyes (63 mm)

cap = cv2.VideoCapture(0)
detector = FaceMeshDetector(maxFaces=1)

while True:
    success, img = cap.read()
    img, faces = detector.findFaceMesh(img)
    
    if faces:
         face = faces[0]
         pointLeft = face[145]
         pointRight = face[374]
         cv2.line(img,pointLeft,pointRight,(0,200,0),3)
         cv2.circle(img,pointLeft,5,(255,0,255),cv2.FILLED)
         cv2.circle(img,pointRight,5,(255,0,255),cv2.FILLED)
         w,_ = detector.findDistance(pointLeft,pointRight)#the underscore makes sure only one value is printed 
         #print(w) # w is the distance in pixels!
         W = 6.3 #cm

         #Finding the focal length
         W = 6.3
         #d = 50 #whatever distance you are away
         #f = (w*d)/W
         #print(f)

    #     #finding the distance now 
    #     ################################################################################################
         f = 634
         d = (W*f)/w
         print(d)
         
         cvzone.putTextRect(img,f'Depth:{int(d)}cm',(face[10][0]-100,face[10][1]-50), scale=2)

    cv2.imshow("Image", img)
    cv2.waitKey(1)


    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
